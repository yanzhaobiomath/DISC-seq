# Figure 5A

```r
library(ggplot2)
library(dplyr)
library(gground)
library(ggprism)
library(tidyverse)
```

```r
GO <- read.delim("GAM_list_macrophage_decontx_GO_enrichment.tsv",sep = " ")
head(GO)
KEGG <- read.delim("GAM_list_macrophage_decontx_kegg_enrichment.tsv",sep = " ")
head(KEGG)
```

```r
use_pathway <- group_by(GO, ONTOLOGY) %>%
  top_n(20, wt = -p.adjust) %>%
  group_by(p.adjust) %>%
  rbind(
    top_n(KEGG,20, -p.adjust) %>%
      group_by(p.adjust) %>%
      mutate(ONTOLOGY ='KEGG')
  ) %>%
  ungroup() %>%
  mutate(ONTOLOGY = factor(ONTOLOGY,
                           levels = rev(c('BP','CC','MF','KEGG')))) %>%
  dplyr::arrange(ONTOLOGY, p.adjust) %>%
  mutate(Description = factor(Description, levels = Description)) %>%
  tibble::rowid_to_column('index')
head(use_pathway)
```

```r
options(repr.plot.width=14, repr.plot.height=18)

pal <- c('#eaa052', '#b74147','#90ad5b','#23929c')

width <-0.65
xaxis_max <- max(-log10(use_pathway$p.adjust)) +3.5
rect.data <- group_by(use_pathway, ONTOLOGY) %>%
  summarize(n = n()) %>%
  ungroup() %>%
  mutate(
    xmin = -3* width,
    xmax = -2* width,
    ymax = cumsum(n),
    ymin = lag(ymax, default =0) +0.6,
    ymax = ymax +0.4
  )
p <- use_pathway %>%
  ggplot(aes(-log10(p.adjust), y = index, fill = ONTOLOGY))+
  geom_round_col(
    aes(y = Description), width =0.5, alpha =0.8
  ) +
  geom_text(
    aes(x =0.05, label = Description),
    hjust =0, size =4.5
  ) +
  geom_text(
    aes(x =0.1, label = geneID, colour = ONTOLOGY),
    hjust =0, vjust =3, size =3, fontface ='italic',
    show.legend =FALSE
  ) +
  geom_point(
    aes(x = -width, size = termRatio),
    shape =21
  ) +
  geom_text(
    aes(x = -width, label = Count ),size = 3.8,fontface ='bold',
  ) +
  scale_size_continuous(name ='Term Ratio', range = c(6,13)) +
  geom_round_rect(
    aes(xmin = xmin, xmax = xmax, ymin = ymin, ymax = ymax,
        fill = ONTOLOGY),
    data = rect.data,
    radius = unit(2,'mm'),
    inherit.aes =FALSE
  ) +
  geom_text(
    aes(x = (xmin + xmax) /2, y = (ymin + ymax) /2, label = ONTOLOGY),fontface ='bold',
    data = rect.data,
    inherit.aes =FALSE
  ) +
  geom_segment(
    aes(x =0, y =0, xend = xaxis_max, yend =0),
    linewidth =1.5,
    inherit.aes =FALSE
  ) +
  labs(y =NULL) +
  scale_fill_manual(name ='Category', values = pal) +
  scale_colour_manual(values = pal) +
  scale_x_continuous(
    breaks = seq(0, xaxis_max,2),
    expand = expansion(c(0,0))
  ) +
  theme_prism() +
  theme(
    axis.text.y = element_blank(),
    axis.line = element_blank(),
    axis.ticks.y = element_blank(),
    legend.title = element_text()
  )
p
ggsave("enrichment_of_macrophage_gam.pdf",width = 14,height = 18,units = "in",dpi = 300)
```

# Figure 5B

```r
library(Seurat)
library(ggplot2)
library(reshape2)
library(patchwork)
```

```r
PBMC <- readRDS("PBMC.Rds")
PBMC
head(PBMC@meta.data)
```

```r
ma_df <- subset(PBMC,subset = anno == "Macrophage")
ma_df <- NormalizeData(ma_df)
```

```r
options(repr.plot.width=5, repr.plot.height=8)
data <- FetchData(ma_df,vars = c("IL1B","NLRP3","CASP1","CASP3","CASP4","CASP8","NOS2",
                                 "NFKB1","NFKB2","RELA","RELB","REL",
                                 "log2ANXV"))
data$log2ANXV <- as.character(data$log2ANXV)
data <- melt(data)
data$log2ANXV <- as.numeric(data$log2ANXV)

mark = 1
for (i in unique(data$variable)){
  data1 <- data[which(data$variable == i),]
  if(mark == 1){
    p <- ggplot(data1,aes(log2ANXV,value))+geom_smooth(method="loess",span = 0.8,se = T)+
 
      theme(panel.grid.major = element_blank(), 
            panel.grid.minor = element_blank(), 
            axis.line = element_line(),
            panel.background = element_rect(fill = 'white'), 
            plot.background=element_rect(fill="white"),
      )+labs(title = i)+ylab("Expression Level")+xlab("log2 ANXV")
    mark = 0 
  }
  else{
    p <- p + ggplot(data1,aes(log2ANXV,value))+geom_smooth(method="loess",span = 0.8,se = T)+
      theme(panel.grid.major = element_blank(), 
            panel.grid.minor = element_blank(),
            axis.line = element_line(),
            panel.background = element_rect(fill = 'white'), 
            plot.background=element_rect(fill="white"),
      )+labs(title = i)+ylab("Expression Level")+xlab("log2 ANXV")
  }
}
p +  plot_layout(ncol = 3)
ggsave("loess_plots.pdf",width = 5,height = 8,units = "in",dpi = 300)
```

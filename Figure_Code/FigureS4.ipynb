```r
options(repr.plot.width=8, repr.plot.height=8)

ext_len1 <- (max(PBMC@reductions$umap@cell.embeddings[,1])-min(PBMC@reductions$umap@cell.embeddings[,1]))/10
ext_len2 <- (max(PBMC@reductions$umap@cell.embeddings[,2])-min(PBMC@reductions$umap@cell.embeddings[,2]))/10
DimPlot(PBMC,group.by="RNA_snn_res.0.5",pt.size = 0.5,raster = F,label = T,repel = T)+
  theme(axis.line = element_blank(),axis.ticks = element_blank(),axis.text = element_blank(),axis.title = element_blank())+
  geom_segment(aes(x = min(PBMC@reductions$umap@cell.embeddings[,1]) , y = min(PBMC@reductions$umap@cell.embeddings[,2]) ,
                   xend = min(PBMC@reductions$umap@cell.embeddings[,1]) +ext_len1, yend = min(PBMC@reductions$umap@cell.embeddings[,2]) ),
               colour = "black", size=1,arrow = arrow(length = unit(0.3,"cm")))+ 
  geom_segment(aes(x = min(PBMC@reductions$umap@cell.embeddings[,1])  , y = min(PBMC@reductions$umap@cell.embeddings[,2])  ,
                   xend = min(PBMC@reductions$umap@cell.embeddings[,1]) , yend = min(PBMC@reductions$umap@cell.embeddings[,2]) + ext_len2),
               colour = "black", size=1,arrow = arrow(length = unit(0.3,"cm"))) +
  annotate("text", x = min(PBMC@reductions$umap@cell.embeddings[,1]) +ext_len1/2, y = min(PBMC@reductions$umap@cell.embeddings[,2]) -ext_len2/3, label = "UMAP_1",
           color="black",size = 4, fontface="bold" ) + 
  annotate("text", x = min(PBMC@reductions$umap@cell.embeddings[,1]) -ext_len1/3, y = min(PBMC@reductions$umap@cell.embeddings[,2]) + ext_len2/2, label = "UMAP_2",
           color="black",size = 4, fontface="bold" ,angle=90)+
  xlim(c(min(PBMC@reductions$umap@cell.embeddings[,1]) -ext_len1/3, max(PBMC@reductions$umap@cell.embeddings[,1])))+
  ylim(c(min(PBMC@reductions$umap@cell.embeddings[,2]) -ext_len2/3,max(PBMC@reductions$umap@cell.embeddings[,2])))
ggsave(filename = "clu_umap.pdf",width = 8,height = 8,units = "in",dpi = 300)
```

# Figure S4D

```r
library(Seurat)
library(ggplot2)
library(reshape2)
```

```r
PBMC <- readRDS("PBMC.Rds")
PBMC
head(PBMC@meta.data)
color <- c("#ea5545", "#27aeef", "#ef9b20", "#b33dc6", "#edbf33", "#87bc45", "#ede15b",  "#0d88e6",
           "#1a53ff", "#bdcf32", "#f46a9b",  "#7c1158", "#b30000","#4421af")
```

```r
options(repr.plot.width=6, repr.plot.height=8)

data <- FetchData(PBMC,vars = c("anno","HTO_classification"))

a <-unique(data$HTO_classification)
b <- unique(data$anno)
result <- data.frame(matrix(ncol = length(b),nrow = 0))
for (i in a) {
  total_frac <- c()
  for (j in b) {
    frac <- length(data$anno[which(data$anno == j & data$HTO_classification == i)])/length(data$HTO_classification[which(data$HTO_classification == i)])
    total_frac <- c(total_frac,frac)
  }
  result <- rbind(result,total_frac)
}
colnames(result) <- b
rownames(result) <- c("Case1","Case2","Case3")
result <- as.data.frame(result)
result$clu <- factor(x=rownames(result),levels = c("Case1","Case2","Case3"),ordered = T)
c <- melt(result)
c$variable <- factor(x = c$variable,levels = unique(data$anno)[order(unique(tolower(data$anno)),decreasing = F)],ordered = T)
ggplot(c,aes(x= clu,y=value,fill = variable))+geom_col()+
  theme(panel.background = element_blank(),axis.text.x = element_text(hjust = 1,angle = 45))+
  ggtitle("Celltype distribution") +labs(x = "Case",y = "Fraction")+
  labs(fill = "Celltypes")+
  scale_fill_manual(values = color)
ggsave(filename = "anno_distribution.pdf",width = 6,height = 8,units = "in",dpi = 300)
```

# Figure S4E

```r
library(Seurat)
library(ggplot2)
library(stringr)
```

```r
PBMC <- readRDS("PBMC.Rds")
PBMC
head(PBMC@meta.data)
```

```r
bubble <- read.delim("annotation_marker_for_bubble_decontx.tsv")
bubble <- bubble[order(bubble$annotation,decreasing = F),]
head(bubble)
```

```r
options(repr.plot.width=25, repr.plot.height=10)

marker <- list()
all <- c()
for(i in unique(bubble$annotation)){
  ma <- str_split(bubble$marker[which(bubble$annotation == i)],pattern = ",",simplify = T)
  marker[[i]] <- unique(c(ma))[which(!(unique(c(ma)) %in% all))]
  all <- c(all,unique(c(ma)))
}
Idents(PBMC) <- PBMC$anno
Idents(PBMC) <- factor(PBMC@active.ident, sort(levels(PBMC@active.ident)))
DotPlot(PBMC, features = marker,cluster.idents = F) +
  scale_color_gradient2(low = "white",mid = "#FA6849",high = "#70020E")+
  RotatedAxis()+
  theme(
    panel.border = element_rect(color="black"), 
    panel.spacing = unit(1, "mm"), 
    strip.text = element_text(margin=margin(b=3, unit="mm"),angle = 90),
    strip.placement = 'outlet', 
    axis.line = element_blank(),
  )+   guides(size = guide_legend(title = "Fraction of cells \n  in group (%)",direction = "horizontal",title.position = "top",label.position = "bottom",keywidth = 1.3),
              color = guide_colorbar(title = "Mean expression \n        in group",ticks.colour = NA,frame.colour = "black",direction = "horizontal",title.position = "top",label.position = "bottom", barwidth = 7))
ggsave("annotation_marker.pdf",width = 25,height = 10,dpi = 300)
```

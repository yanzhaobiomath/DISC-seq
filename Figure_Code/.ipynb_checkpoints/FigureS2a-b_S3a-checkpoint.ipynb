```r
runGAM <- function (obj, #input object
                    name, #name of sample 
                    outdir = './', 
                    n = 2000, # number of most significant genes to shown on heatmap
                    n.feature = 2000,
                    width = 8, height = 15){
    
    require(Seurat)
    require(scran)
    require(tradeSeq)
    require(clusterProfiler)
    require(org.Hs.eg.db)
    
    if (class(obj) == 'Seurat'){
    
        features = VariableFeatures(obj) ## remove this later
        bg = rownames(obj)[rowSums(obj)>0]
        obj = as.SingleCellExperiment(obj)
        
        }else if (class(obj) == 'SingleCellExperiment'){

        obj = scran::modelGeneVar(obj)
        features = scran::getTopHVGs(obj, n = n.feature)
        bg = rownames(obj)[rowSums(counts(obj))>0]
        
        }else{
            stop('obj should be either Seurat or obj')
        }
    
    subobj = obj[features,]
    coldf = colData(subobj)
    subobj <- fitGAM(counts(subobj), 
                  pseudotime = subobj$log2ANXV, 
                  cellWeights = matrix(1, nrow = ncol(subobj), ncol = 1))
    ATres <- associationTest(subobj)
    
    colData(subobj) = cbind(coldf, colData(subobj))

    # saveRDS(subobj, file.path(outdir, paste0('obj_GAM_', name, '.rds')))
    write.csv(ATres, file.path(outdir, paste0('GAM_AssTest_', name, '.csv')))
    
    topgenes <- rownames(ATres[order(ATres$pvalue), ])[1:n]
    pst.ord <- order(subobj$log2ANXV, na.last = NA)
    heatdata <- assays(subobj)$counts[topgenes, pst.ord]
    heatdata <- log2(heatdata + 1)
    heatclus <- subobj$HTO_classification[pst.ord]
    names(heatclus) = colnames(heatdata)

    anno_gene <- ATres$pvalue[order(ATres$pvalue)][1:n]
    names(anno_gene) <- topgenes
    # conpal = RColorBrewer::brewer.pal("PuBu", n = 9)
    # genecol = setNames(colpal[length(topgenes)],  topgenes)

    samples = unique(heatclus)
    # mycol = dichromat(colorschemes$BluetoOrange.12)
    # mycol = setNames(RColorBrewer::brewer.pal(name = "Set2", n = length(samples)),  samples)
    colpal = RColorBrewer::brewer.pal(name = "Set3", n = 12)
    mycol = setNames(colpal[1:length(samples)],  samples)

    filename = file.path(outdir, paste0('GAM_', name, '.pdf'))

    
    p = Heatmap(heatdata,  cluster_columns = F, show_column_names = F, 
            # col = colorRamp2(c(0, 5, 10), c("blue", "white", "yellow")),
            # col = mycol,
            name = 'Exp',
            top_annotation = HeatmapAnnotation(
                sample = heatclus,
                annotation_legend_param = list(sample = list(title = "sample")),
                col = list(sample = mycol)
            ),
               left_annotation = rowAnnotation(
                   p.val = anno_gene,
                   annotation_legend_param = list(p.val = list(title = "p.val"))
               # col = list(p.val = genecol)
               ))


    pdf(filename, width = width, height = height)
    draw(p)
    dev.off()

    sigenes = rownames(ATres[ATres$pvalue < 0.05 &(!is.na(ATres$pvalue)),])
    
    
    
    gene.df <- bitr(sigenes, fromType = "SYMBOL",
            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
            OrgDb = org.Hs.eg.db)
    
    bg.df <- bitr(bg, fromType = "SYMBOL",
            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
            OrgDb = org.Hs.eg.db)
    

    runGoKEGG(gene.df = gene.df, bg.df = bg.df, name = name, outdir = outdir)

    # return(ATres)
    
    }
```

```r
# remotes::install_github("YuLab-SMU/clusterProfiler") 
# # install the packages
# remotes::install_github("YuLab-SMU/createKEGGdb")
# # import the library and create a KEGG database locally 
# library(createKEGGdb)
# species <-c("ath","hsa","mmu", "rno","dre","dme","cel")
# species <- c('hsa')
# createKEGGdb::create_kegg_db(species = species)
# You will get KEGG.db_1.0.tar.gz file in your working directory

install.packages("KEGG.db_1.0.tar.gz", repos=NULL,type="source")
library(KEGG.db)



runGoKEGG <-function(gene.df, bg.df = NULL, name, outdir = "./",
                                          top = 10,
                    width = 7, height =5, charLength = 60, kegg_organism = 'hsa', GO_orgdb = 'org.Hs.eg.db'){

    if (is.null(bg.df)){
        gopath = enrichGO(gene = gene.df$SYMBOL, 
                      ont = 'ALL',
                      OrgDb = 'org.Hs.eg.db', 
                      keyType = 'SYMBOL',
                      universe = bg.df)
    }else{
        gopath = enrichGO(gene = gene.df$SYMBOL, 
                      ont = 'ALL',
                      OrgDb = 'org.Hs.eg.db', 
                      keyType = 'SYMBOL',
                      universe = bg.df$SYMBOL)
        }
    godf = data.frame(gopath)
    main = name
    if (!is.null(godf) & (nrow(godf) > 0)  & (rownames(godf)[1] != "NA")){
        # godf = gopath@result
        print(nrow(godf))
        godf$Sample = name
        godf$termRatioStr =paste0(godf$Count,'/', unlist(strsplit(godf$BgRatio, '/'))[seq(1,nrow(godf)*2,2)])
          godf$termRatio = round(godf$Count/as.numeric(unlist(strsplit(godf$BgRatio, '/'))[seq(1,nrow(godf)*2,2)]),2)
          godf = godf[order(godf$pvalue, decreasing=T),]
        write.csv(godf, file.path(outdir, paste0('GO_', name, '.csv')))
        if (nrow(godf)>top){
            godf = godf[(nrow(godf)-top+1):nrow(godf),]
          }
                
          ## Normalize term description ##
          terms = as.character(godf$Description)
        print(head(terms))
          terms = lapply(terms, function(x,k){
            x = as.character(x)
            if(nchar(x)>k){x=substr(x,start=1,stop=k)}
            return(x)}, charLength)
          godf$Description = do.call(rbind, terms)
          godf = godf[!duplicated(godf$Description),]
          
         
          godf = godf[order(godf$p.adjust, decreasing=T),]
          p3 <- ggplot(godf, aes(x=-log10( p.adjust ), y=Description)) +
            geom_point(aes( size= termRatio, colour = Count)  ) + scale_y_discrete(limits=godf$Description)+
            ggtitle(paste("GO enrichment for", main))  +  scale_color_gradient(low = 'blue', high = 'red') + #xlim(range(log10(out$pvalue))) +
            theme(axis.text.x=element_text(angle=0, size=10, vjust=0.7), 
                  axis.text.y=element_text(angle=0,size=12, vjust=0.7),
                  plot.title = element_text(lineheight=.8, face="bold", hjust=0, size =12),
                  panel.background = element_rect(fill="white", colour='gray'),
                  panel.grid.major = element_line(size = 0.05, colour = "gray"),
                  panel.grid.minor.y = element_line(size=0.05, colour="gray"),
                  panel.grid.minor.x = element_line(size=0.05, colour="gray")
            )
          # ggsave(file.path(outdir, paste0(main,"group_GO_bar.pdf")), p1, width = width, height=height)
          # ggsave(file.path(outdir,paste0(main,"group_Go_Bubble.pdf")), p2, width = width, height=height)
          ggsave(file.path(outdir,paste0(main,"group_Go_Bubble_ty2_y-padj.pdf")), p3, width = width, height=height)
              ggsave(file.path(outdir,paste0(main,"group_Go_Bubble_ty2_y-padj.png")), p3, width = width, height=height)

    }
    
    
    
}
```

```r
runGAM_genecl <- function (obj, #input object
                    name, #name of sample 
                    outdir = './', 
                    n = 2000, # number of most significant genes to shown on heatmap
                    n.feature = 2000,
                    row_km = 5,
                    width = 8, height = 15){
    
    # require(Seurat)
    require(scran)
    require(tradeSeq)
    require(clusterProfiler)
    require(org.Hs.eg.db)
    
    if (class(obj) == 'Seurat'){
    
        features = VariableFeatures(obj) ## remove this later
        bg = rownames(obj)[rowSums(obj)>0]
        obj = as.SingleCellExperiment(obj)
        
        }else if (class(obj) == 'SingleCellExperiment'){

        obj = scran::modelGeneVar(obj)
        features = scran::getTopHVGs(obj, n = n.feature)
        bg = rownames(obj)[rowSums(counts(obj))>0]
        
        }else{
            stop('obj should be either Seurat or obj')
        }
    
    subobj = obj[features,]
    coldf = SingleCellExperiment::colData(subobj)
    subobj <- tradeSeq::fitGAM(SingleCellExperiment::counts(subobj), 
                  pseudotime = subobj$log2ANXV, 
                  cellWeights = matrix(1, nrow = ncol(subobj), ncol = 1))
    ATres <- tradeSeq::associationTest(subobj)
    
    SummarizedExperiment::colData(subobj) = cbind(coldf, SummarizedExperiment::colData(subobj))

    # saveRDS(subobj, file.path(outdir, paste0('obj_GAM_', name, '.rds')))
    write.csv(ATres, file.path(outdir, paste0('GAM_AssTest_', name, '.csv')))

    ATres = ATres[ATres$pvalue < 0.05 &(!is.na(ATres$pvalue)),]
    print(dim(ATres))

    if (n < nrow(ATres)){
        
        topgenes <- rownames(ATres[order(ATres$pvalue), ])[1:n]
        ATres <- ATres[order(ATres$pvalue), ][1:n,]
        
        }else{
        
        topgenes <- rownames(ATres)
        n = nrow(ATres)
        
        }
    
    pst.ord <- order(subobj$log2ANXV, na.last = NA)
    heatdata <- SummarizedExperiment::assays(subobj)$counts[topgenes, pst.ord]
    heatdata <- log2(heatdata + 1)
    heatclus <- subobj$HTO_classification[pst.ord]
    names(heatclus) = colnames(heatdata)

    anno_gene <- ATres$pvalue[order(ATres$pvalue)]
    names(anno_gene) <- topgenes
    # conpal = RColorBrewer::brewer.pal("PuBu", n = 9)
    # genecol = setNames(colpal[length(topgenes)],  topgenes)

    set.seed(6)
    samples = unique(heatclus)
    # mycol = dichromat(colorschemes$BluetoOrange.12)
    # mycol = setNames(RColorBrewer::brewer.pal(name = "Set2", n = length(samples)),  samples)
    colpal = RColorBrewer::brewer.pal(name = "Set3", n = 12)
    mycol = setNames(colpal[1:length(samples)],  samples)
    
    heat_cluster = subobj$seurat_clusters[pst.ord]
    clusters = unique(heat_cluster)
    clust_col = setNames(colpal[1:length(clusters)],  clusters)
    log2ANXV = subobj$log2ANXV[pst.ord]

    # genecol <- colorRampPalette(c("white", colpal[1]))(length(anno_gene))
    # genecol <- setNames(genecol, anno_gene)
    genecol <- colorRampPalette(c("white", colpal[1]))(100)

    filename = file.path(outdir, paste0('GAM_', name, '.pdf'))

   p = Heatmap(heatdata,  cluster_columns = F, show_column_names = F, 
           row_km = row_km,
            name = 'Exp',
            top_annotation = HeatmapAnnotation(
                sample = heatclus,
                cluster = heat_cluster,
                log2ANXV = anno_barplot(log2ANXV, gp = gpar(fill = 3, col = 3)),
                # annotation_legend_param = list(sample = list(title = "sample")),
                col = list(sample = mycol, cluster =  clust_col)
            ),
               left_annotation = rowAnnotation(
                   p.val = anno_gene,
                  annotation_legend_param = list(p.val = list(title = "p.val"))
                   # col = list(p.val = colpal[1])
                   # show_legend = FALSE
               
               ))


    pdf(filename, width = width, height = height)
    draw(p)
    dev.off()

    pdf(filename, width = width, height = height)
    draw(p)
    dev.off()

    
    
    gene.df <- bitr(rownames(heatdata), fromType = "SYMBOL",
            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
            OrgDb = org.Hs.eg.db)
    
    bg.df <- bitr(bg, fromType = "SYMBOL",
            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
            OrgDb = org.Hs.eg.db)


    clustlist = row_order(draw(p))
    saveRDS(clustlist, file.path(outdir, paste0("heatmap_geneclust_", name, ".rds")))
    
    for(i in names(clustlist)){
        print(i)
        idx = clustlist[[i]]
        # genes = rownames(heatdata)[idx]
        runGoKEGG(gene.df = gene.df[idx,], bg.df = bg.df, name = paste0(name, "_genecluster_", i), outdir = outdir)
        
    }

    return(p)
    
    }
```

```r
DEG_bycluster <- function(seu, name = name, outdir = "./", logfc.threshold = 0.1){
    
    Idents(seu) = seu$seurat_clusters
    # markerdf = FindAllMarkers(seu, logfc.threshold = logfc.threshold)
    features = VariableFeatures(seu)

    clusters = unique(seu$seurat_clusters)
    # print(head(markerdf))
    # if (!is.null(markerdf)){
        for (i in 1:length(clusters)){
            if (i != length(clusters)){
            for (j in (i+1):length(clusters)){

                x = clusters[i]
                y = clusters[j]
                
                markerdf = FindMarkers(seu, logfc.threshold = logfc.threshold,  ident.1 = x, ident.2 = y)
                markerdf$gene = rownames(markerdf)
                write.csv(markerdf, file.path(outdir, paste0(name,'_cluster_',x, '_vs_cluster_', y, '_markers.csv')))
                                # print(markerdf)
                    
                if (!is.null(markerdf)){
                    
                    upgenes = markerdf$gene[markerdf$avg_log2FC > 0 &(markerdf$p_val_adj < 0.05)]
                    downgenes = markerdf$gene[markerdf$avg_log2FC < 0 &(markerdf$p_val_adj < 0.05)]
                    print(length(downgenes))
            
                    bg = rownames(seu)[rowSums(seu)>0]
                
                    upgene.df <- bitr(upgenes, fromType = "SYMBOL",
                            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
                            OrgDb = org.Hs.eg.db)
                     upgene.df =  upgene.df[!is.na( upgene.df$ENTREZID),]
            
                    downgene.df <- bitr(downgenes, fromType = "SYMBOL",
                            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
                            OrgDb = org.Hs.eg.db)
                    downgene.df =  downgene.df[!is.na(downgene.df$ENTREZID),]
                    
                    bg.df <- bitr(bg, fromType = "SYMBOL",
                            toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
                            OrgDb = org.Hs.eg.db)
                     bg.df =  bg.df[!is.na(bg.df$ENTREZID),]
        
            
                    runGoKEGG(gene.df = upgene.df, bg.df = bg.df, name = paste0(name, '_cluster_',x, '_vs_cluster_', y, "_up_", i), outdir = outdir)
                    runGoKEGG(gene.df = downgene.df, bg.df = bg.df, name = paste0(name, '_cluster_',x, '_vs_cluster_', y, "_down_", i), outdir = outdir)
                    
                }
                }
            }
        }
    
}
```

```r
DEG_bycluster_all <- function(seu, name = name, outdir = "./", logfc.threshold = 0.1){
    
    Idents(seu) = seu$seurat_clusters
    markerdf = FindAllMarkers(seu, logfc.threshold = logfc.threshold)
    features = VariableFeatures(seu)

    write.csv(markerdf, file.path(outdir, paste0(name, '_markers.csv')))

    clusters = unique(seu$seurat_clusters)
    # print(head(markerdf))
    if (!is.null(markerdf)){
        for (i in clusters){

            upgenes = markerdf$gene[markerdf$avg_log2FC > 0 &(markerdf$cluster == i) &(markerdf$p_val_adj < 0.05)]
            downgenes = markerdf$gene[markerdf$avg_log2FC < 0 &(markerdf$cluster == i)&(markerdf$p_val_adj < 0.05)]
            print(length(downgenes))
    
            bg = rownames(seu)[rowSums(seu)>0]
        
            upgene.df <- bitr(upgenes, fromType = "SYMBOL",
                    toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
                    OrgDb = org.Hs.eg.db)
             upgene.df =  upgene.df[!is.na( upgene.df$ENTREZID),]
    
            downgene.df <- bitr(downgenes, fromType = "SYMBOL",
                    toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
                    OrgDb = org.Hs.eg.db)
            downgene.df =  downgene.df[!is.na(downgene.df$ENTREZID),]
            
            bg.df <- bitr(bg, fromType = "SYMBOL",
                    toType = c("ENSEMBL", "SYMBOL", "ENTREZID"),
                    OrgDb = org.Hs.eg.db)
             bg.df =  bg.df[!is.na(bg.df$ENTREZID),]

    
            runGoKEGG(gene.df = upgene.df, bg.df = bg.df, name = paste0(name, "_up_", i), outdir = outdir)
            runGoKEGG(gene.df = downgene.df, bg.df = bg.df, name = paste0(name, "_down_", i), outdir = outdir)
            
        }
        }
    
}
```
